```
import matplotlib.pyplot as plt 
import numpy as np

plot_files = ['train_cost.npy', 'train_acc.npy', 'train_mIOU.npy']
plot_names = ['Cost', 'Global Accuracy', 'Mean Intersection over Union']

experiment_list = ['Batch_size/1', 'Batch_size/2', 'Batch_size/4']
for i_experiment in experiment_list:
  plt.figure()
  for idx, i_file in enumerate(plot_files):
      plot_path = os.path.join('./', 'Experiments', i_experiment, 'Model/Data', i_file)
      train_x = np.load(plot_path, allow_pickle=True, encoding='latin1')
      num_epochs = list(range(1, len(train_x)+1))
      plt.subplot(1, 3, idx+1)
      plt.plot(num_epochs, train_x, lw=2)
      plt.ylabel(plot_names[idx])
      plt.xlabel('Epoch')
plt.rcParams['figure.figsize'] = (20,3)
plt.show()
```
